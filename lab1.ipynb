#### 1. Create a random 4x4 array from an exponential distribution. Refer numpy functions.

```python
rng = np.random.default_rng(1)
```

```python
rand_exp = rng.exponential(scale=1, size=(4,4))
rand_exp
```

#### 2. Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better.

```python
exp_vector = rng.exponential(scale=1, size=(10000))
```

```python
plt.hist(
    np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="Uniform"
)

plt.hist(
    np.random.randn(100000), density=True, bins=100, histtype="step", color="green", label="Normal"
)

plt.hist(
    exp_vector, density=True, bins=100, histtype="step", color="red", label="Exponential"
)

plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

plt.show()
```

#### 3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]. 


```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = (X**2 + Y**2)

figure = plt.figure(1, figsize = (12, 4))
subplot3d = plt.subplot(111, projection='3d')
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, linewidth=0.1)
plt.show()
```

```python
df = pd.read_csv('data/Pokemon.csv', index_col=0, encoding='latin')
```

```python
new_df = df[['HP', 'Attack','Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
new_df.head()
```

```python
pearson_corr_matrix = new_df.corr(method='pearson')
pearson_corr_matrix
spearman_corr_matrix = new_df.corr(method='spearman')
spearman_corr_matrix
```

#### 4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.

```python
plt.figure(1)
plt.title('Pearson standard coefficient')
sns.heatmap(pearson_corr_matrix, cmap='twilight')
plt.show()
```

```python
plt.figure(1)
plt.title('Spearman rank correlation')
sns.heatmap(spearman_corr_matrix, cmap='twilight')
plt.show()
```

## Task 2

```python
nano_df =pd.read_csv('data/Au_nanoparticle_dataset.csv', index_col=0, encoding='latin')
```

#### 1. Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.

```python
filter = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
filtered_df = nano_df[filter]
```

#### 2. Display the first 20 samples of this dataframe.

```python
filtered_df.head(20)
```

#### 3. Calculate the mean, standard deviation and quartile values for each of the above 4 features.

```python
mean = filtered_df.mean()
std = filtered_df.std()
quartile = filtered_df.quantile([0.25, 0.75])

print(f'Mean values: \n{mean}\n')
print(f'Standard deviation: \n{std}\n')
print(f'Quartiles: \n{quartile}\n')
```

#### 4. Plot the histogram of each of these features in a 1x4 layout.

```python
for column in filtered_df.columns:
    plt.hist(filtered_df[column], bins=30, label=column, color='blue')
    plt.title(column)
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

```

#### 5. Visualize the scatter plots and histograms of this dataframe using the pairplot functionality of seaborn library.

```python
sns.pairplot(filtered_df)
plt.suptitle('Filtered data pairplot')
plt.show()
```

#### 6. a. Plots on the diagonal contains the histogram of each feature along with the kernal density estimation plot. 

#### b. Plots on the lower half [g.map_lower] contains the bivariate kernal density estimation plot.

```python
g = sns.PairGrid(filtered_df)

g.map_upper(sns.histplot) 
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)

plt.show()
```
